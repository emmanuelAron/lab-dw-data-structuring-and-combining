```python
# Your code goes here
import pandas as pd
df1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
```

```python
#Columns definition
cols = ["customer" ,"state" , "gender" , "education", "customer lifetime value" , 
        "income" ,"monthly premium auto" , "number of open complaints" ,"policy type" , "vehicle class" , "total claim amount"]

#Names in lowercase
df1.columns = df1.columns.str.lower()
df2.columns = df2.columns.str.lower()
df3.columns = df3.columns.str.lower()
df1 = df1.rename(columns={'st': 'state'})
df2 = df2.rename(columns={'st': 'state'})

df1 = df1[cols]
df2 = df2[cols]
df3 = df3[cols]

display('df1 ',df1.head())
display('df2 ',df2.head())
display('df3' ,df3.head())



```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
pd.set_option('display.max_columns', None)
df4 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
df4 = df4.rename(columns={
    'customer_lifetime_value': 'customer lifetime value',
    'total_claim_amount': 'total claim amount',
    'number_of_policies': 'number of open complaints',
    'policy_type': 'policy type',
    'vehicle_class': 'vehicle class'
})
# Delete the index in double
df4.drop(columns=['unnamed:_0'], inplace=True)

#Reorder the df4 columns the same way than the others
df4 = df4[[col for col in cols if col in df4.columns] + 
          [col for col in df4.columns if col not in cols]]

df4.columns = df4.columns.str.replace('_', ' ')


df4.head()
```
